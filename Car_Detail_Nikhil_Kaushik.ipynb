# **Key Findings**

🚗 General Trends
Most cars are manually driven and sold by individual sellers rather than dealers.

First-owner cars dominate the dataset, indicating that people tend to sell early in the ownership cycle.

The dataset covers a wide range of car ages, from newer models (e.g., 2017+) to older ones (2005+).

💰Selling Price Insights
Selling price is inversely related to car age — newer cars sell for more.

Cars with fewer kilometers driven tend to have higher resale values.

Diesel vehicles generally have higher average selling prices than petrol ones.

Automatic transmission cars may command higher prices but are fewer in number.

🛢️ Fuel Type Patterns
Petrol cars are more common, but diesel cars show higher average prices.

CNG and LPG cars, if present, tend to be older and lower-priced (in larger datasets).

👥 Ownership and Seller Impact
First-owner cars retain value better than second or third-owner cars.

Dealer sales are less frequent but may show slightly higher pricing.

Individual sellers make up the majority of transactions.

📈 Trend Analysis
Average car price decreases steadily with each passing year of the vehicle's age.

Outliers exist in both selling_price and km_driven, especially in luxury or premium vehicles.

There’s a moderate correlation between km_driven and selling price (negatively).
